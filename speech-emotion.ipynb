```python
import librosa
import numpy as np
import pandas as pd
import glob
import os

base_dir = '/kaggle/input/speech-data-for-emotions'
all_data = []

# --- CREMA-D ---
crema_emotions = {
    'ANG': 'angry', 'DIS': 'disgust', 'FEA': 'fearful', 'HAP': 'happy',
    'NEU': 'neutral', 'SAD': 'sad'
}
crema_dir = os.path.join(base_dir, 'crema-d-mirror-main-AudioWAV', 'crema-d-mirror-main-AudioWAV', 'AudioWAV')
for path in glob.glob(crema_dir + '/*.wav'):
    fname = os.path.basename(path)
    emotion_code = fname.split('_')[2]
    if emotion_code in crema_emotions:
        emotion = crema_emotions[emotion_code]
        try:
            y, sr = librosa.load(path, sr=16000)
            mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
            mfcc = np.mean(mfcc.T, axis=0)
            all_data.append({'features': mfcc, 'label': emotion, 'source': 'crema-d'})
        except Exception as e:
            print(f"Skipped {path}: {e}")

# --- RAVDESS ---
ravdess_emotions = {
    '01': 'neutral', '02': 'calm', '03': 'happy', '04': 'sad',
    '05': 'angry', '06': 'fearful', '07': 'disgust', '08': 'surprised'
}
ravdess_dir = os.path.join(base_dir, 'Audio_Speech_Actors_01-24')
for path in glob.glob(ravdess_dir + '/Actor_*/*.wav'):
    fname = os.path.basename(path)
    emotion_code = fname.split('-')[2]
    if emotion_code in ravdess_emotions:
        emotion = ravdess_emotions[emotion_code]
        try:
            y, sr = librosa.load(path, sr=16000)
            mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
            mfcc = np.mean(mfcc.T, axis=0)
            all_data.append({'features': mfcc, 'label': emotion, 'source': 'ravdess'})
        except Exception as e:
            print(f"Skipped {path}: {e}")

# --- TESS ---
tess_dir = os.path.join(base_dir, 'archive', 'TESS Toronto emotional speech set data')
for emotion_folder in os.listdir(tess_dir):
    if emotion_folder.startswith('.'): continue
    emotion = emotion_folder.split('_')[-1].lower()
    for path in glob.glob(os.path.join(tess_dir, emotion_folder, '*.wav')):
        try:
            y, sr = librosa.load(path, sr=16000)
            mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
            mfcc = np.mean(mfcc.T, axis=0)
            all_data.append({'features': mfcc, 'label': emotion, 'source': 'tess'})
        except Exception as e:
            print(f"Skipped {path}: {e}")

# Convert to DataFrame
df = pd.DataFrame(all_data)
print(df.head())
print(df['label'].value_counts())
```

```python
# --- 1. Imports ---
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# --- 2. Dataset class ---
class SpeechEmotionDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.long)
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# --- 3. Split data ---
dataset = SpeechEmotionDataset(X, y)
val_size = int(0.15 * len(dataset))
test_size = int(0.10 * len(dataset))
train_size = len(dataset) - val_size - test_size
train_ds, val_ds, test_ds = random_split(dataset, [train_size, val_size, test_size], generator=torch.Generator().manual_seed(42))

batch_size = 64
train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=batch_size)
test_loader = DataLoader(test_ds, batch_size=batch_size)

# --- 4. CNN Model ---
class DeeperCNN(nn.Module):
    def __init__(self, input_dim, num_classes):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(1, 32, kernel_size=3, padding=1), nn.BatchNorm1d(32), nn.ReLU(), nn.Dropout(0.3),
            nn.Conv1d(32, 64, kernel_size=3, padding=1), nn.BatchNorm1d(64), nn.ReLU(), nn.Dropout(0.3),
            nn.Conv1d(64, 128, kernel_size=3, padding=1), nn.BatchNorm1d(128), nn.ReLU(), nn.Dropout(0.3),
            nn.AdaptiveAvgPool1d(1)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128, 64), nn.ReLU(), nn.Dropout(0.3),
            nn.Linear(64, num_classes)
        )
    def forward(self, x):
        x = x.unsqueeze(1) # [B, 1, input_dim]
        x = self.net(x)
        x = self.classifier(x)
        return x

input_dim = X.shape[1]
num_classes = len(np.unique(y))
model = DeeperCNN(input_dim, num_classes).to("cuda" if torch.cuda.is_available() else "cpu")
device = next(model.parameters()).device

# --- 5. Training setup ---
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
num_epochs = 40
early_stop_patience = 8

train_accs, val_accs = [], []
train_losses, val_losses = [], []
best_val_acc = 0
patience = 0

# --- 6. Training Loop ---
for epoch in range(1, num_epochs+1):
    model.train()
    tr_loss, tr_correct, tr_total = 0, 0, 0
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        out = model(xb)
        loss = loss_fn(out, yb)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        tr_loss += loss.item() * xb.size(0)
        tr_correct += (out.argmax(1) == yb).sum().item()
        tr_total += xb.size(0)
    train_losses.append(tr_loss/tr_total)
    train_accs.append(tr_correct/tr_total)

    # Validation
    model.eval()
    val_loss, val_correct, val_total = 0, 0, 0
    with torch.no_grad():
        for xb, yb in val_loader:
            xb, yb = xb.to(device), yb.to(device)
            out = model(xb)
            loss = loss_fn(out, yb)
            val_loss += loss.item() * xb.size(0)
            val_correct += (out.argmax(1) == yb).sum().item()
            val_total += xb.size(0)
    val_losses.append(val_loss/val_total)
    val_acc = val_correct/val_total
    val_accs.append(val_acc)

    print(f"Epoch {epoch}: train_loss={train_losses[-1]:.3f}, train_acc={train_accs[-1]:.3f}, val_loss={val_losses[-1]:.3f}, val_acc={val_acc:.3f} [Best: {best_val_acc:.3f}]")

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), "best_model.pt")
        patience = 0
    else:
        patience += 1
        if patience > early_stop_patience:
            print("Early stopping!")
            break

print(f"Training done! Best val_acc = {best_val_acc:.3f}")

# --- 7. Plot accuracy/loss curves ---
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(train_accs, label="Train Acc")
plt.plot(val_accs, label="Val Acc")
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.legend()
plt.subplot(1,2,2)
plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Val Loss")
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.legend()
plt.suptitle("Training Curves")
plt.show()

# --- 8. Confusion Matrix on Test Set ---
model.load_state_dict(torch.load("best_model.pt"))
model.eval()
all_preds, all_trues = [], []
with torch.no_grad():
    for xb, yb in test_loader:
        xb = xb.to(device)
        out = model(xb)
        preds = out.argmax(1).cpu().numpy()
        all_preds.extend(preds)
        all_trues.extend(yb.numpy())
cm = confusion_matrix(all_trues, all_preds)
disp = ConfusionMatrixDisplay(cm, display_labels=sorted(label_to_idx, key=label_to_idx.get))
disp.plot(xticks_rotation=45)
plt.title("Confusion Matrix (Test Set)")
plt.show()
```

```python
# --- 1. Imports ---
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# --- 2. Data Preparation ---
# Load your dataframe here (assume it's named df and has 'features', 'label')
# Example: df = pd.read_pickle("features_labels.pkl")

# Label encode
emotion_labels = sorted(df['label'].unique())
label_to_idx = {label: idx for idx, label in enumerate(emotion_labels)}
idx_to_label = {idx: label for label, idx in label_to_idx.items()}

# Convert features/labels
X = np.stack(df['features'].to_numpy())
y = np.array([label_to_idx[label] for label in df['label']])

# --- 3. Dataset and DataLoader ---
class SpeechEmotionDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.long)

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

dataset = SpeechEmotionDataset(X, y)

# --- 4. Split Data ---
# 70% train, 15% val, 15% test
n_total = len(dataset)
n_train = int(0.7 * n_total)
n_val = int(0.15 * n_total)
n_test = n_total - n_train - n_val
train_set, val_set, test_set = random_split(dataset, [n_train, n_val, n_test], generator=torch.Generator().manual_seed(42))
batch_size = 64
train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True, drop_last=True)
val_loader = DataLoader(val_set, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_set, batch_size=batch_size, shuffle=False)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- 5. CNN + BiLSTM Model ---
import torch.nn.functional as F

class CNN_BiLSTM(nn.Module):
    def __init__(self, num_classes, feature_dim, lstm_hidden=128, lstm_layers=1, dropout=0.3):
        super(CNN_BiLSTM, self).__init__()
        self.conv1 = nn.Conv1d(1, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv1d(32, 64, kernel_size=3, padding=1)
        self.pool = nn.MaxPool1d(2)
        self.dropout = nn.Dropout(dropout)
        # Figure out LSTM input dimension after convolutions/pooling
        with torch.no_grad():
            dummy = torch.zeros(1, 1, feature_dim)
            x = self.conv1(dummy)
            x = F.relu(x)
            x = self.pool(x)
            x = self.dropout(x)
            x = self.conv2(x)
            x = F.relu(x)
            x = self.pool(x)
            x = self.dropout(x)
            conv_out_dim = x.shape[1]
            conv_out_len = x.shape[2]
        self.conv_out_dim = conv_out_dim
        self.conv_out_len = conv_out_len

        self.lstm = nn.LSTM(input_size=conv_out_dim, hidden_size=lstm_hidden, num_layers=lstm_layers,
                            batch_first=True, bidirectional=True, dropout=dropout)
        self.fc1 = nn.Linear(2*lstm_hidden, 128)
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = x.unsqueeze(1)   # (batch, 1, features)
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        x = self.dropout(x)
        x = F.relu(self.conv2(x))
        x = self.pool(x)
        x = self.dropout(x)
        x = x.permute(0, 2, 1)  # (batch, seq_len, channels)
        lstm_out, _ = self.lstm(x)
        out = lstm_out[:, -1, :]   # last time step
        out = torch.relu(self.fc1(out))
        out = self.dropout(out)
        out = self.fc2(out)
        return out

num_classes = len(emotion_labels)
feature_dim = X.shape[1]
model = CNN_BiLSTM(num_classes, feature_dim, lstm_hidden=128, lstm_layers=2, dropout=0.3).to(device)

# # To use DeeperCNN instead, comment above and uncomment below:
# model = DeeperCNN(num_classes, feature_dim).to(device)

# --- 6. Training ---
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

def train_one_epoch(loader, train=True):
    if train:
        model.train()
    else:
        model.eval()
    total_loss, correct, count = 0, 0, 0
    for xb, yb in loader:
        xb, yb = xb.to(device), yb.to(device)
        if train:
            optimizer.zero_grad()
        out = model(xb)
        loss = criterion(out, yb)
        if train:
            loss.backward()
            optimizer.step()
        total_loss += loss.item() * xb.size(0)
        preds = torch.argmax(out, 1)
        correct += (preds == yb).sum().item()
        count += xb.size(0)
    return total_loss / count, correct / count

num_epochs = 50
train_losses, val_losses, train_accs, val_accs = [], [], [], []
best_val_acc = 0

for epoch in range(num_epochs):
    tr_loss, tr_acc = train_one_epoch(train_loader, train=True)
    va_loss, va_acc = train_one_epoch(val_loader, train=False)
    train_losses.append(tr_loss)
    val_losses.append(va_loss)
    train_accs.append(tr_acc)
    val_accs.append(va_acc)
    if va_acc > best_val_acc:
        best_val_acc = va_acc
    print(f"Epoch {epoch+1}: train_loss={tr_loss:.3f}, train_acc={tr_acc:.3f}, val_loss={va_loss:.3f}, val_acc={va_acc:.3f} [Best: {best_val_acc:.3f}]")

# --- 7. Plot Accuracy/Loss ---
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Val Loss')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.subplot(1,2,2)
plt.plot(train_accs, label='Train Acc')
plt.plot(val_accs, label='Val Acc')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.tight_layout()
plt.show()

# --- 8. Confusion Matrix ---
model.eval()
all_preds, all_trues = [], []
with torch.no_grad():
    for xb, yb in test_loader:
        xb, yb = xb.to(device), yb.to(device)
        out = model(xb)
        preds = torch.argmax(out, 1)
        all_preds.extend(preds.cpu().numpy())
        all_trues.extend(yb.cpu().numpy())

cm = confusion_matrix(all_trues, all_preds)
disp = ConfusionMatrixDisplay(cm, display_labels=[idx_to_label[i] for i in range(len(idx_to_label))])
fig, ax = plt.subplots(figsize=(8,8))
disp.plot(xticks_rotation=45, ax=ax)
plt.title("Confusion Matrix (Test Set)")
plt.show()

# --- 9. Bar Chart: Per-Class Accuracy ---
import seaborn as sns
all_preds, all_trues = np.array(all_preds), np.array(all_trues)
class_accs = []
for i in range(num_classes):
    mask = all_trues == i
    acc = (all_preds[mask] == i).sum() / mask.sum()
    class_accs.append(acc)
plt.figure(figsize=(8,5))
sns.barplot(x=[idx_to_label[i] for i in range(num_classes)], y=class_accs)
plt.ylabel("Per-class Accuracy")
plt.ylim(0,1)
plt.title("Per-class Accuracy (Test Set)")
plt.show()
# Save model weights to a .pt file
torch.save(model.state_dict(), 'emotion_cnn_weights.pt')
```
